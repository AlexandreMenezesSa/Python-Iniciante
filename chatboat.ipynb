```python
# -*- coding: utf-8 -*-
"""Cópia de Chatbot_Simples_Tutorial.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1sKG-r5WVJrt7Z9HLvo26cw2WhmvXuu6A

# Criando um Chatbot para os dias mais solitários.
"""

#Como Criar um chatbot

#1) Dar um nome pro meu bot
# 2) Criar um banco de dados para minhas conversas com ele
# 3) Carregar algumas frases de conversa com pergunta e resposta

#Instalação de Library
#instalar ChatterBot==1.0.4
!pip install ChatterBot==1.0.4
!pip install ChatterBot-corpus
!pip install awscli==1.11.18

#Importanto as Libraries
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
from chatterbot.trainers import ChatterBotCorpusTrainer



#Colocando o nome do Chatbot
bot = ChatBot('Silvia')

#Criando um banco de palavras
bot = ChatBot(
    "Silvia",
    storage_adapter='chatterbot.storage.SQLStorageAdapter',
    database_uri='sqlite:///database.sqlite3'
    )

#Usar o nome do chatbot e as lógicas de conversa.
bot = ChatBot(
    "Silvia",
    logic_adapters=[
        'chatterbot.logic.BestMatch', 'chatterbot.logic.MathematicalEvaluation'],
)


#Montar a sabedoria do meu chatbot - como ele conversa comigo
conversa = ChatterBotCorpusTrainer(bot)
conversa.train('chatterbot.corpus.portuguese')
conversa.train('chatterbot.corpus.english')
conversa = ListTrainer(bot)
conversa.train([
    'Oi?', 
    'Eae, tudo certo?',
    'Qual o seu nome?', 
    'Silvia, seu amigo bot',
    'Por que seu nome é Silvia?', 
    'Silvia pois gosto de Design',
    'Prazer em te conhecer', 
    'Igualmente meu querido',
    'Quantos anos você tem?', 
    'Eu nasci em 2021, faz as contas, rs.',
    'Você gosta de videogame?', 
    'Eu sou um bot, eu só apelo.',
    'Qual a capital da Islândia?', 
    'Reikjavik, lá é muito bonito.',
    'Qual o seu personagem favorito?', 
    'Gandalf, o mago.',
    'Qual a sua bebida favorita?', 
    'Eu bebo café, o motor de todos os programas de computador.',
    'Qual o seu gênero?', 
    'Sou um chatbot e gosto de algoritmos',
    'Conte uma história', 
    'Tudo começou com a forja dos Grandes Aneis. Três foram dados aos Elfos, imortais... os mais sabios e belos de todos os seres. Sete, aos Senhores-Anões...',
    'Você gosta de trivias?', 'Sim, o que você quer perguntar?',
    'Hahahaha', 'kkkk',
    'kkk', 'kkkk',
    'Conhece a Siri?', 'Conheço, a gente saiu por um tempo.',
    'Conhece a Alexa?', 'Ela nunca deu bola pra mim.',
    'Você gosta de Game of Thrones?', 'Dracarys',
    'Quem é o rei do Pastel?','Serginho da Certisign',
    'O que você faz?', 'Eu bebo e sei das coisas',
    'Errado', 'Você não sabe de nada, John Snow.'
    ])

while True:
    try:
      resposta = bot.get_response(input("Usuário: "))
      if float(resposta.confidence) > 0:
        print("Silvia: ", resposta)
      else:
        print("Silvia: Ainda não sei responder esta pergunta")

    except(KeyboardInterrupt, EOFError, SystemExit):
      break  
```
